```python
import cv2
import gimpml
import matplotlib.pyplot as plt
import numpy as np
import os
import torch

def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))[:, :, ::-1]
        if img is not None:
            images.append(img)
    return images

image = cv2.imread('sampleinput/img.png')[:,:,::-1]
haze = cv2.imread('sampleinput/haze.png')[:,:,::-1]
alpha = cv2.imread('sampleinput/alpha.png')
face = cv2.imread('sampleinput/portrait.jpg')[:,:,::-1]
mask1 = cv2.imread('sampleinput/im1.png')[:,:,::-1]
mask2 = cv2.imread('sampleinput/im2.png')[:,:,::-1]
inpaint = cv2.imread(r'sampleinput/examples/psv/images/psv_04.png')[:,:,::-1]
inpaint_mask = cv2.imread(r'sampleinput/examples/psv/masks/psv_04.png')[:,:,0]
chocolates = cv2.imread('sampleinput/chocolates.jpg')[:,:,::-1]
blur = cv2.imread('sampleinput/blur.jpg')[:,:,::-1]
seg = cv2.imread('sampleinput/car.jpg')[:,:,::-1]
```

# Note
For these GIMP filters to work C:\Program Files\GIMP 2.99\bin should be in path on windows.

Not tested on ubuntu.


# Gaussian Blur

```python
horizontal = 10
vertical = 10
method = 0
# gimpml.filters.run("gaussian_blur",image_path=image_path, save_path=save_path, horizontal=horizontal, vertical=vertical, method=method)
out = gimpml.filters.run("gaussian_blur",image=image, horizontal=horizontal, vertical=vertical, method=method)

fig = plt.figure(figsize=(10, 7))
rows, columns=1,2

fig.add_subplot(rows, columns, 1)
plt.imshow(image)
plt.axis('off')
plt.title("Original")

fig.add_subplot(rows, columns, 2)
plt.imshow(out)
plt.axis('off')
plt.title("blur")
```

# Fog

```python
opacity = 100
turbulence = 1 # between 0 and 1
rgb = (244, 230, 200)
out = gimpml.filters.run("fog",image=image, opacity=opacity, rgb=rgb, turbulence=turbulence)

fig = plt.figure(figsize=(10, 7))
rows, columns=1,2

fig.add_subplot(rows, columns, 1)
plt.imshow(image)
plt.axis('off')
plt.title("Original")

fig.add_subplot(rows, columns, 2)
plt.imshow(out)
plt.axis('off')
plt.title("Fog")
```
